```python
original_data['CASE_NO_letter'] = original_data['CASE_NO'].str.extract(r'([A-Za-z]+)')
original_data['CASE_NO_number'] = original_data['CASE_NO'].str.extract(r'([0-9]+)')
```

```python
import re
```

```python
def extract_letter_number (case_no) :
    letter_pattern = re.compile(r'[A-Za-z]+')
    number_pattern = re.compile(r'[0-9]+')
    
    letter = letter_pattern.search(case_no).group() # 查一下group的功能
    number = number_pattern.search(case_no).group()
    
    return (letter, number)
```

```python
extract_letter_number('CI00222')
```

```python
original_data['CASE_NO'].apply(extract_letter_number)
```

```python
def cyr2wyr (year) :
    wyr = int(year) + 1911
    return wyr
```

```python
original_data['Year'].apply(cyr2wyr)
```

```python
original_data['WestYear'] = original_data['Year'].apply(cyr2wyr)
```
